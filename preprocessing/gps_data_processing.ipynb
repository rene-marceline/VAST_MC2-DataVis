```python
# Storing stop points 
# Apr 19, 2023, Jisoo Lee

import os
import pandas as pd
import csv
from datetime import datetime, timedelta
import numpy as np

folder_path = "./data/gps"  

# create the "GG" folder if it doesn't exist
output_path = "./data/stop_points_fixed"  
if not os.path.exists(output_path):
    os.makedirs(output_path)

all_data = pd.DataFrame()  # create an empty DataFrame to store all the data

time_threshold = 60
time_thresholds = [60, 300, 1800, 3600, 7200, 10800, 86400]
#threshold: 1min, 5min, 30min, 1hour, 2hour, 3hour, more than 3 hours 

for file_name in os.listdir(folder_path):
    if file_name.endswith(".csv"):
        file_path = os.path.join(folder_path, file_name)
        file_name_only = os.path.splitext(os.path.basename(file_path))[0]
        print("Loading data from file:", file_path)
        
        new_data = [['id', 'long1', 'long2', 'lat1', 'lat2', 'long_center', 'lat_center', 'Timestamp1', 'Timestamp2', 'time_threshold']]
        for j in range(len(time_thresholds)-1):
 
            file = open(file_path, "r")
            data = list(csv.reader(file, delimiter=","))
            file.close()
            

            count = 0
            for i in range(2, len(data)):
                previous = datetime.strptime(data[i-1][0], '%m/%d/%Y %H:%M:%S')
                current = datetime.strptime(data[i][0], '%m/%d/%Y %H:%M:%S')

                time_difference = current-previous
                
                long_center = str(np.mean([float(data[i-1][3]), float(data[i][3])])) 
                lat_center = str(np.mean([float(data[i-1][2]), float(data[i][2])])) 
                # only check time gap within a same day. 
                if (time_difference>timedelta(seconds=time_thresholds[j]) and time_difference<timedelta(seconds=time_thresholds[j+1])and current.date()-previous.date()<timedelta(seconds=1)):
                    count = count+1
#                     print(data[i][1], data[i-1][2], data[i][2], data[i-1][3], data[i][3], data[i-1][0], data[i][0], time_thresholds[j])
                    temp = [data[i][1], data[i-1][3], data[i][3], data[i-1][2], data[i][2], long_center, lat_center, data[i-1][0], data[i][0], time_thresholds[j]]
                    new_data.append(temp)
        
        print("Saving {}".format(f'{output_path}/{file_name_only}_stoppoints.csv'))
        with open(f'{output_path}/{file_name_only}_stoppoints.csv', mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(new_data)

         
        
```

```python
# merge stop points with shop name with card data(cc, lc)

# load stoppoints with shop name file
stop_file = open('./data/merged_stoppoints_shop2.csv', "r")
stop_data = list(csv.reader(stop_file, delimiter=","))
stop_file.close()

print(stop_data[0])
print(stop_data[1])
```

```python
cc_file = open('./data/cc_data.csv', "r")
cc_data = list(csv.reader(cc_file, delimiter=","))
cc_file.close()
cc_data[0] = ['timestamp', 'location', 'price', 'last4ccnum']
```

```python
import time
import datetime

# add cc data to stop data
merge_cc = stop_data.copy()
merge_cc[0].append('cc')

for i in range(1, len(merge_cc)):
    stamp1 = time.mktime(datetime.datetime.strptime(merge_cc[i][-4],"%m/%d/%Y %H:%M:%S").timetuple())
    stamp2 = time.mktime(datetime.datetime.strptime(merge_cc[i][-3],"%m/%d/%Y %H:%M:%S").timetuple())
    merge_cc[i].append([])
    for j in range(1, len(cc_data)):
        if merge_cc[i][-2]==cc_data[j][1]:
            cstamp = time.mktime(datetime.datetime.strptime(cc_data[j][0],"%m/%d/%Y %H:%M").timetuple())
            if stamp1<cstamp and cstamp<stamp2:
                merge_cc[i][-1].append([cc_data[j][0], cc_data[j][-1], cc_data[j][2]])
        
print(merge_cc[0])
print(merge_cc[1])
```

```python
lc_file = open('./data/loyalty_data.csv', "r")
lc_data = list(csv.reader(lc_file, delimiter=","))
lc_file.close()
lc_data[0] = ['timestamp', 'location', 'price', 'loyaltynum']
```

```python
import time
import datetime
from datetime import datetime

merge_lc = merge_cc.copy()
merge_lc[0].append('lc')

for i in range(1, len(merge_lc)):
    date1 = datetime.strptime(merge_lc[i][-5], '%m/%d/%Y %H:%M:%S')
    merge_lc[i].append([])
    for j in range(1, len(lc_data)):    
        if merge_lc[i][-3]==lc_data[j][1]:
            date2= datetime.strptime(lc_data[j][0], '%m/%d/%Y')
            if date1.date()==date2.date():
                merge_lc[i][-1].append([lc_data[j][0], lc_data[j][-1], lc_data[j][2]])

        
print(merge_lc[0])
print(merge_lc[1])
```

```python
with open('./data/merged_stoppoints_shop_cards_v2.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(merge_lc)
```

```python
id_name_file = open('./data/car-assignments.csv', "r")
id_name_file_data = list(csv.reader(id_name_file, delimiter=","))
id_name_file.close()

id_name_file_data[0] = ['LastName', 'FirstName','CarID','CurrentEmploymentType','CurrentEmploymentTitle']

# create dictionary of employee name and car id
name_dict = {}
for i in range(1,len(id_name_file_data)):
    name = id_name_file_data[i][0] +' '+id_name_file_data[i][1]
    if "" in name_dict:
        name_dict[""] = name_dict[""] +' or '+name
#         print(name_dict[""])
    else:
        name_dict[id_name_file_data[i][2]] = name

print(name_dict)
```

```python
truck_names = name_dict[""]
del name_dict[""]

#assign truck driver names to trucks
for i in (["101", "104", "105", "106", "107"]):
    name_dict[i] = truck_names

print(name_dict)
```
